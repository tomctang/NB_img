```python
counter = 1
def make_plot(fn=1, xmin=-1, xmax=1, ymin=-1, ymax=1, lineslope=1, lineyint=0, figname="plot"):
    global counter
    savename = figname + str(counter) + ".pdf"
    counter = counter + 1
    FIGSIZEX = 3
    FIGSIZEY = 3
    
    xs = np.linspace(xmin, xmax)
    ys = np.array(list(map(fn, xs)))
    lineys = [lineyint + lineslope*x for x in xs]
    
    plt.plot(xs, ys, color='k')
    plt.plot(xs, lineys, color='k')
    
    axes = plt.gca()
    axes.spines['top'].set_visible(False)
    axes.spines['right'].set_visible(False)
    #axes.spines['left'].set_position('zero')
    #axes.spines['bottom'].set_position('zero')
    xticks = [-10 + .5*x for x in range(41)]
    plt.xticks(xticks)
    plt.yticks(xticks)
    plt.xlim([xmin, xmax])
    plt.ylim([ymin, ymax])
    plt.xlabel("x")
    plt.ylabel("y")
    axes.grid(True, linestyle='--')
    
    fig = plt.gcf()
    fig.set_size_inches(FIGSIZEX, FIGSIZEY)
    fig.savefig(savename, bbox_inches='tight')
```

```python

def make_one_plot(fn=1, xmin=-1, xmax=1, ymin=-1, ymax=1, figname="plot", clist=None):
    global counter
    savename = figname + str(counter) + ".pdf"
    counter = counter + 1
    FIGSIZEX = 3
    FIGSIZEY = 3

    xs = np.linspace(xmin, xmax)
    if not clist:
        clist = [-1 + x*1./3 for x in range(7)] 
    for c in clist:
        f = lambda x: fn(x, c)
        ys = np.array(list(map(f, xs)))
        plt.plot(xs, ys, color='k')
    
    axes = plt.gca()
    axes.spines['top'].set_visible(False)
    axes.spines['right'].set_visible(False)
    #axes.spines['left'].set_position('zero')
    #axes.spines['bottom'].set_position('zero')
    #xticks = [-10 + .5*x for x in range(41)]
    xticks = [-10 + x for x in range(21)]
    yticks = [-20 + x for x in range(41)]
    plt.xticks(xticks)
    plt.yticks(yticks)
    plt.xlim([xmin, xmax])
    plt.ylim([ymin, ymax])
    plt.xlabel("x")
    plt.ylabel("y")
    axes.grid(True, linestyle='--')
    
    fig = plt.gcf()
    fig.set_size_inches(FIGSIZEX, FIGSIZEY)
    fig.savefig(savename, bbox_inches='tight')
```

```python
def diff(x,y):
    return x/y # try also x+y

x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)

# use x,y
for j in x:
    for k in y:
        slope = diff(j,k)
        domain = np.linspace(j-0.07,j+0.07,2)
        def fun(x1,y1):
            z = slope*(domain-x1)+y1
            return z
        plt.plot(domain,fun(j,k),solid_capstyle='projecting',solid_joinstyle='bevel')
```

```python

def make_slopefield(fn=1, xmin=-3, xmax=3, ymin=-2, ymax=2, scale=12, figname="plot"):
    global counter
    savename = figname + str(counter) + ".pdf"
    counter = counter + 1
    FIGSIZEX = 3
    FIGSIZEY = 3

    xs = [-2,-1, 0, 1, 2]
    ys = [-1, 0, 1]
   
    us = []
    vs = []
    for y in ys:
        urow = []
        vrow = []
        for x in xs:
            slope = fn(x,y)
            normalizer = (slope**2 + 1)**.5
            xcomp = 1/normalizer
            ycomp = slope/normalizer
            urow.append(xcomp)
            vrow.append(ycomp)
        us.append(urow)
        vs.append(vrow)
            
    plt.quiver(xs, ys, us, vs, headwidth=1, headlength=0, scale=scale)
    
    axes = plt.gca()
    axes.spines['top'].set_visible(False)
    axes.spines['right'].set_visible(False)
    #axes.spines['left'].set_position('zero')
    #axes.spines['bottom'].set_position('zero')
    #xticks = [-10 + .5*x for x in range(41)]
    xticks = [-10 + x for x in range(21)]
    yticks = [-20 + x for x in range(41)]
    plt.xticks(xticks)
    plt.yticks(yticks)
    plt.xlim([xmin, xmax])
    plt.ylim([ymin, ymax])
    plt.xlabel("x")
    plt.ylabel("y")
    #axes.grid(True, linestyle='--')
    
    fig = plt.gcf()
    fig.set_size_inches(FIGSIZEX, FIGSIZEY)
    fig.savefig(savename, bbox_inches='tight')
```

```python
counter = 1
def f(x,y):
    return y**2 - x
make_slopefield(f, scale=10)
```
