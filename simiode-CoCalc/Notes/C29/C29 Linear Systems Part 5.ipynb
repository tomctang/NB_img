# C29 Linear Systems Part 5

Here is a catalog of solution types for systems of two first-order, linear differential equations with constant coefficients:

$\qquad \begin{bmatrix} x' \\ y' \end{bmatrix} = \begin{bmatrix*}[r] a & b \\ c & d \end{bmatrix*} \begin{bmatrix} x \\ y \end{bmatrix}$

with (for the specific solution) initial conditions

$\qquad \begin{bmatrix} x(0) \\ y(0) \end{bmatrix} = \begin{bmatrix} x_0 \\ y_0 \end{bmatrix}$.

The phase plane is plotted on the domain $-l \leq x \leq l$ and $-l \leq y \leq l$.  The $x' = 0$ nullcline is red, the $y' = 0$ nullcline is blue, the specific solution is magenta, and other solutions are green.

## Example 1: Stable Node

From Part 1.

Problem: $\begin{bmatrix} x' \\ y' \end{bmatrix} = \begin{bmatrix*}[r] -5 & 3 \\ 1 & -3 \end{bmatrix*} \begin{bmatrix} x \\ y \end{bmatrix} $.

Solution: $
\begin{bmatrix} x \\ y \end{bmatrix} = 
a\begin{bmatrix} 1 \\ 1 \end{bmatrix}e^{-2t} + 
b\begin{bmatrix*}[r] 3 \\ -1 \end{bmatrix*}e^{-4t}
$

For initial conditions $x(0) = 0$ and $y(0) = 4$, the coefficients are $a = 3$ and $b = -1$.

```python
A = matrix([[-5, 3], [1, -3]]); (x0, y0) = (0, 4); l = 4
var('x y'); (xp, yp) = A*vector([x, y])
xnull = implicit_plot(xp == 0, (x,-l,l), (y,-l,l), color='red')
ynull = implicit_plot(yp == 0, (x,-l,l), (y,-l,l), color='blue')
ics = [(1,1), (-1,-1), (3,-1), (-3,1), (0,2), (0,-2), (0,-4), (3,0), (-3,0)]
solns = streamline_plot([xp, yp], (x,-l,l), (y,-l,l), start_points = ics, color='green')
ssoln = streamline_plot([xp, yp], (x,-l,l), (y,-l,l), start_points = [(x0, y0)], color='magenta')
show(xnull + ynull + solns + ssoln, aspect_ratio=1)
```

```python
var('t')
x = function('x')(t)
y = function('y')(t)
(xp, yp) = A*vector([x, y])
de1 = diff(x, t) == xp
de2 = diff(y, t) == yp
gsoln = desolve_system([de1, de2], [x,y], ivar = t); show(gsoln)
ssoln = desolve_system([de1, de2], [x,y], ivar = t, ics = (0, x0, y0)); show(ssoln)
```
